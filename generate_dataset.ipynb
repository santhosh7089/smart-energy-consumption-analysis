# Creating data Sets

```python
import pandas as pd
import numpy as np

# -----------------------------
# Configuration
# -----------------------------
np.random.seed(42)

students = {
    "Student_A": "Room-1",
    "Student_B": "Room-2",
    "Student_C": "Room-3",
    "Student_D": "Room-4"
}

appliances = {
    "AC": (1.2, 2.0),
    "Fan": (0.05, 0.2),
    "Light": (0.02, 0.1),
    "Laptop": (0.05, 0.15),
    "Refrigerator": (0.1, 0.3),
    "Heater": (1.0, 2.5)
}

date_range = pd.date_range(
    start="2024-01-01",
    end="2024-01-31 23:00",
    freq="H"
)

# -----------------------------
# Generate Data
# -----------------------------
data = []

for timestamp in date_range:
    hour = timestamp.hour

    for student, room in students.items():
        for appliance, (low, high) in appliances.items():

            energy = np.random.uniform(low, high)

            # Night-time wastage patterns
            if appliance == "AC" and hour >= 22:
                energy *= 1.8

            if appliance == "Heater" and hour >= 21:
                energy *= 2.0

            if appliance in ["Light", "Fan"] and hour < 6:
                energy *= 0.3

            # Random anomaly
            if np.random.rand() < 0.01:
                energy *= 3

            data.append([
                timestamp,
                room,
                student,
                appliance,
                round(energy, 3)
            ])

# -----------------------------
# Create DataFrame
# -----------------------------
columns = ["timestamp", "room", "student", "appliance", "energy_kwh"]
df = pd.DataFrame(data, columns=columns)

# -----------------------------
# Save CSV in same folder
# -----------------------------
df.to_csv("generate_data.csv", index=False)

# Preview
df.head()
```
