# STEP-4: FEATURE ENGINEERING

```python
import pandas as pd
import numpy as np
print("All libraries imported successfully")
```

```python
df = pd.read_csv("generate_data.csv")
print("Data sets loded successfully")
df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
print("Converts string time into datetime format")
```

```python
df['hour'] = df['timestamp'].dt.hour
print("Hours extract from time")
```

```python
df['is_night'] = df['hour'].apply(
    lambda x: 1 if (x >= 23 or x <= 5) else 0
    )
print("Night time column created from hour (1 = night, 0 = day)")
```

```python
threshold = df['energy_kwh'].mean()

df['high_consumption'] = df['energy_kwh'].apply(
    lambda x: 1 if x > threshold else 0
)
print("High consumption column created from energy_kwh (1 = high, 0 = low)")

```

```python
df['wastage_flag'] = (
    (df['is_night'] == 1) &
    (df['high_consumption'] == 1)
).astype(int)
print("Wastage flag column created from night and high consumption (1 = wastage, 0 = normal)")
```

```python
appliance_mean = df.groupby('appliance')['energy_kwh'].transform('mean')
df['efficiency_score'] = df['energy_kwh'] / appliance_mean
print("Efficiency score column created by comparing energy usage with appliance average")
```

```python
df['is_peak_hour'] = df['hour'].apply(lambda x: 1 if 17 <= x <= 21 else 0)
print("Peak hour column created from hour (1 = peak, 0 = non-peak)")
```

```python
print("Displaying first 5 rows of all key columns to check feature creation")
df[['timestamp','appliance','energy_kwh','is_night','is_peak_hour','high_consumption','wastage_flag','efficiency_score']].head()
```

## INSIGHTS FROM FEATURE ENGINEERING :

#### 1. Night-time high consumption is a major contributor to energy wastage.
#### 2. Peak-hour usage (6 PM – 10 PM) shows significantly higher energy consumption.
#### 3. Wastage flag identifies exact appliances and time periods causing inefficiency.
#### 4. Appliance efficiency scores highlight devices consuming unusually high energy relative to their type.
#### 5. Rule-based features make the dataset ML-ready and decision-ready.
#### 6. Student-wise and room-wise patterns quantify individual contributions to wastage.
#### 7. The dataset is transformed from raw readings to actionable intelligence for cost-saving and optimization.
