# STEP-5: ANOMALY DETECTION

```python
import pandas as pd
import numpy as np
print("Libraries imported successfully")
```

```python
df = pd.read_csv("generate_data.csv")
df['timestamp'] = pd.to_datetime(df['timestamp'])
print("Data set loaded successfully")
df.head()
```

```python
mean_energy = df['energy_kwh'].mean()
std_energy = df['energy_kwh'].std()
print("Mean of the energy_kwh:",mean_energy)
print("STD of the energy_kwh:",std_energy)
```

```python
# We consider energy usage as anomalous if:
# - energy_kwh > mean + 2 × standard deviation

df['stat_anomaly'] = df['energy_kwh'].apply(
    lambda x: 1 if x > (mean_energy + 2 * std_energy) else 0
)
```

```python
# If energy consumption > threshold → wastage
df['wastage_flag'] = np.where(df['energy_kwh'] > 5, 1, 0)
```

```python
# Efficiency = usage compared to average
df['efficiency_score'] = df['energy_kwh'] / df['energy_kwh'].mean()
```

```python
df['final_anomaly'] = (
    (df['wastage_flag'] == 1) |
    (df['stat_anomaly'] == 1) |
    (df['efficiency_score'] > 1.5)
).astype(int)

print("Final anomaly flag created using rule-based and statistical conditions.")
```

```python
anomalies_sample = df[df['final_anomaly'] == 1].head()

print("Displaying sample records identified as anomalies:")
anomalies_sample
```

```python
appliance_anomalies = df[df['final_anomaly'] == 1].groupby('appliance')['energy_kwh'].count()

print("Calculated number of anomalies per appliance:")
appliance_anomalies.sort_values(ascending=False)
```

```python
student_anomalies = df[df['final_anomaly'] == 1].groupby('student')['energy_kwh'].count()

print("Calculated number of anomalies per student:")
student_anomalies.sort_values(ascending=False)
```

```python
room_anomalies = df[df['final_anomaly'] == 1].groupby('room')['energy_kwh'].count()

print("Calculated number of anomalies per room:")
room_anomalies.sort_values(ascending=False)
```

```python
hourly_anomalies = df[df['final_anomaly'] == 1].groupby(df['timestamp'].dt.hour)['energy_kwh'].count()

print("Calculated number of anomalies occurring at each hour of the day:")
hourly_anomalies
```

```python
total_anomalies = df['final_anomaly'].sum()

print("STEP-5 ANOMALY DETECTION COMPLETED")
print("Total number of anomalous energy usage records detected:", total_anomalies)
print(df.columns)
```

## INSIGHTS FROM ANOMALY DETECTION (STEP-5):

#### 1. Night-time energy usage accounts for a significant portion of anomalies.
#### 2. AC and Heater appliances are the most frequently inefficient devices.
#### 3. Certain students consistently contribute to abnormal energy consumption.
#### 4. Specific rooms show repeated anomalous behavior, indicating appliance or usage issues.
#### 5. Rule-based and statistical methods effectively detect wastage without machine learning.
