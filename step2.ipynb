# STEP 2: DATA UNDERSTANDING & VALIDATION

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("generate_data.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
df["hour"] = df["timestamp"].dt.hour
df["day"] = df["timestamp"].dt.day
df["weekday"] = df["timestamp"].dt.day_name()

df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
print("Appliance usage by high to low")
appliance_usage = df.groupby("appliance")["energy_kwh"].mean().sort_values(ascending=False)
appliance_usage
```

```python
print("Student consumption by high to low")
student_usage = df.groupby("student")["energy_kwh"].mean().sort_values(ascending=False)
student_usage
```

```python
print("Room consumption by high to low")
room_usage = df.groupby("room")["energy_kwh"].mean().sort_values(ascending=False)
room_usage
```

```python
print("Time Wasteage")
avg_energy = df["energy_kwh"].mean()

wastage_df = df[
    ((df["hour"] >= 22) | (df["hour"] <= 5)) &
    (df["appliance"].isin(["AC", "Heater"])) &
    (df["energy_kwh"] > avg_energy)
]

wastage_df.head()
```

```python
print("Wastage Records")
len(wastage_df)
```

```python
df.to_csv("clean_energy_data.csv", index=False)
wastage_df.to_csv("wastage_records.csv", index=False)
print("DATA UNDERSTANDING & VALIDATION COMPLETED")
```
