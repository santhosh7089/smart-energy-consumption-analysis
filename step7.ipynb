```python
import pandas as pd

print("STEP 7: Creating final dataset for Power BI")

# 1. LOAD YOUR DATASET (change filename if needed)
df = pd.read_csv("generate_data.csv")
print("Dataset loaded successfully")

# 2. CREATE FINAL DATAFRAME
final_df = df.copy()

# 3. CREATE PREDICTED ENERGY COLUMN (from ML step)
if 'predicted_energy' not in final_df.columns:
    final_df['predicted_energy'] = final_df['energy_kwh']
    print("predicted_energy column added")

# 4. CREATE WASTAGE FLAG
if 'wastage_flag' not in final_df.columns:
    avg_energy = final_df['energy_kwh'].mean()
    final_df['wastage_flag'] = final_df['energy_kwh'] > avg_energy
    print("wastage_flag column added")

# 5. CREATE EFFICIENCY SCORE
if 'efficiency_score' not in final_df.columns:
    final_df['efficiency_score'] = round(1 / (final_df['energy_kwh'] + 0.01), 3)
    print("efficiency_score column added")

# 6. SAVE FINAL CSV FOR POWER BI
final_df.to_csv("final_energy_analysis.csv", index=False)

print("✅ final_energy_analysis.csv created successfully")
print("➡️ Ready to import into Power BI")
```
